```python
text = """
Natural Language Processing enables computers to understand human languages.
Processing text data includes tokenization, stemming, and lemmatization.
"""
```

```python
tokens = word_tokenize(text)
print("TOKENS:")
print(tokens)
print()
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

stemmer = PorterStemmer()
stemmed_words = []

for word in tokens:
    if word.isalpha():  # remove punctuation
        stemmed_words.append(stemmer.stem(word.lower()))
```

```python
print("STEMMED WORDS:")
print(stemmed_words)
print()
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = []

for word in tokens:
    if word.isalpha():  # remove punctuation
        lemmatized_words.append(lemmatizer.lemmatize(word.lower()))
```

```python
print("LEMMATIZED WORDS:")
print(lemmatized_words)
```
