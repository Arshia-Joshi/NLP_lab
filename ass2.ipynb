```python
texts = """
    Natural language processing is interesting,
    In LAB 5311LA
"""
tokens = word_tokenize(texts)
pos_tags = pos_tag(tokens)

print(f"\nText: {texts}")
print("POS Tags:")
for word, tag in pos_tags:
    print(f"{word:15} {tag}")
```
