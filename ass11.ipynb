```python
import nltk
from nltk import word_tokenize, pos_tag, sent_tokenize
from nltk.chunk import RegexpParser
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
text = """
Artificial intelligence and machine learning are rapidly transforming the global technology landscape.
Many innovative companies are developing advanced deep learning models to solve complex real-world problems.
The growing demand for intelligent systems in healthcare, finance, and transportation is driving significant research and development efforts.
Researchers and engineers are focusing on natural language processing, computer vision, and robotics to build efficient and scalable solutions.
"""
```

```python
grammar = "NP: {<DT>?<JJ>*<NN.*>+}"
chunk_parser = RegexpParser(grammar)
```

```python
noun_phrases = []

sentences = sent_tokenize(text)

for sentence in sentences:
    words = word_tokenize(sentence)
    tagged = pos_tag(words)

    tree = chunk_parser.parse(tagged)


    for subtree in tree.subtrees():
        if subtree.label() == 'NP':
            np = " ".join(word for word, tag in subtree.leaves())
            noun_phrases.append(np)
```

```python
print("Extracted Noun Phrases:\n")

for np in noun_phrases:
    print("-", np)
```

```python
unique_nps = list(set(noun_phrases))

print("Unique Noun Phrases:\n")
for np in unique_nps:
    print("-", np)
```
